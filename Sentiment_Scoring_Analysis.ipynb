### Testing Vader

```python
import nltk
#nltk.download('vader_lexicon')
from nltk.sentiment import SentimentAnalyzer
from nltk import tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sentiment = SentimentAnalyzer()

#polarity score values
#['neg', 'neu', 'pos', 'compound']

sentiment = list()

neg=list()
neu=list()
pos=list()
compound=list()

for sentence in sentences:
    sent_vader = list(SentimentIntensityAnalyzer().polarity_scores(sentence).values())
    neg.append(sent_vader[0])
    neu.append(sent_vader[1])
    pos.append(sent_vader[2])
    compound.append(sent_vader[3])
    

vader_sentences = pd.DataFrame({'sentence':sentences,'neg':neg,'pos':pos,'neu':neu,'compound':compound})
vader_sentences

```

### Testing TextBlob

```python
from textblob import TextBlob

polarity = list()
subjectivity = list()


for sentence in sentences:
    sent = TextBlob(sentence)
    #print(sentence ,' ', sent.sentiment)
    polarity.append(sent.sentiment.polarity)
    subjectivity.append(sent.sentiment.subjectivity)
    sent.sentiment.subjectivity
    
textblob_sentences = pd.DataFrame({'sentence':sentences,'polarity':polarity,'subjectivity':subjectivity})
textblob_sentences
    

    
```

### Testing Flair

```python
from flair.models import TextClassifier
from flair.data import Sentence

classifier = TextClassifier.load('en-sentiment')
#classifier = TextClassifier.load('sentiment-fast')
#classifier = TextClassifier.load('sentiment')

value = list()
score = list()


for sentence in sentences:
    sent = Sentence(sentence)
    classifier.predict(sent)
    value.append(sent.labels[0].value)
    score.append(sent.labels[0].score)
    #print('Sentence polarity: ', sent.labels)
    

    
flair_sentences = pd.DataFrame({'sentence':sentences,'value':value,'score':score})
flair_sentences
    
```

### Convert Sentiment Score for NEGATIVE  Value

```python
value = list()
score = list()

sentences = ['Not ready for promotion','Teamwork!']

for sentence in sentences:
    sent = Sentence(sentence)
    classifier.predict(sent)
    
    if sent.labels[0].value == 'NEGATIVE':
                sent.labels[0].score = 0 - sent.labels[0].score
    else:
                sent.labels[0].score = sent.labels[0].score
            
    value.append(sent.labels[0].value)
    score.append(sent.labels[0].score)
            
new_flair_sentences = pd.DataFrame({'sentence':sentences,'value':value,'score':score})
new_flair_sentences         
```
